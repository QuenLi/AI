```python
%matplotlib inline # 經典命令，會讓jupyter可以直接呈現圖像，並且保存下來。

import numpy as np
import matplotlib.pyplot as plt
```

MatLab 是一個古老的數學計算繪圖的強大軟件，因此python社群的人就想把它放在python環境下使用，這就是matplotlib的由來。以下為相當於 `Hello world`的示範:

```python
x = np.arange(0, 6, 0.1)
y = np.sin(x)

plt.plot(x, y)
plt.show()
```

這個庫這麼樣的強大，可見並不會只做這麼簡單的繪圖，以下進一步示範:

```python
x = np.arange(0, 6, 0.1)
y1 = np.sin(x)
y2 = np.cos(x)

# 在一張座標圖上畫兩條線
plt.plot(x, y1, label='sin')
plt.plot(x, y2, label='cos', linestyle='--')
plt.xlabel('x')
plt.ylabel('y')
plt.title('sin & cos')
plt.legend() # 顯示一個圖例
plt.show()
```

這些庫可以很好的做搭配，因為底下其實都是`numpy`在做事。

```python
x = np.arange(0, 3 * np.pi, 0.1)
y1 = np.sin(x)
y2 = np.cos(x)

plt.subplot(2, 1, 1) # subplot 一共有兩張，上下各有一張圖, 最末尾表示先處理第一張圖
plt.plot(x, y1)
plt.title('sine')

plt.subplot(2, 1, 2)
plt.plot(x, y2)
plt.title('cosine')

plt.show()
```

要把這個概念學明白學清楚要從哪裡開始呢?
先從一個概念開始，Axe。這並不是斧頭的意思，而是表示一張圖所有關鍵元素合起來叫做 Axes。有一篇官方的[舊版文檔](https://matplotlib.org/2.0.2/faq/usage_faq.html)裏頭有提到一張圖如下:

<img src="assets/Axe_photo.webp">

```python
x = np.arange(0, 20, 0.1)
y1 = x**0.5
y2 = np.sin(x)
y3 = np.cos(x)
```

```python
fig, axes = plt.subplots(2, 1, figsize=(14, 12)) # 還有其他參數，參閱matplot官方文檔

# 第一張圖畫兩條線
axes[0].plot(x, y1, label='sqrt')
axes[0].plot(x, y2, label='sin')

axes[0].set_title('sqrt(x) vs sin(x)', fontsize=18)
axes[0].legend()

axes[1].plot(x, y1, label='sqrt')
axes[1].plot(x, y3, label='cos')

axes[1].set_title('sqrt(x) vs cos(x)', fontsize=18)
axes[1].legend()

# 下面我們在進一步訂製
axes[1].set_xlabel('x axis', fontsize=14, fontfamily='serif', fontstyle='italic')
axes[1].set_ylabel('y axis', fontsize='x-large', fontstyle='oblique')

axes[1].set_aspect('equal') # x, y 軸的尺寸範圍一樣而不會自動伸縮

axes[1].set_xlim(0, 16)

start, end = axes[1].get_xlim()
axes[1].xaxis.set_ticks(np.arange(start, end, 1))

axes[1].xaxis.set_tick_params(rotation=45, labelsize='large', color='b')
axes[1].yaxis.tick_right()

axes[1].minorticks_on()

axes[1].grid(which='major', axis='both', color='r')
axes[1].grid(which='minor', axis='both')

for ax in axes:
    ax.set_anchor('W') # W 代表 west，向左邊對齊

plt.subplots_adjust(hspace=0.3) # 兩副圖高度的垂直方向間隔設定
plt.show()
```

```python
x = np.arange(0, 20, 0.1)
y1 = x**0.5
y2 = np.sin(x)
y3 = np.cos(x)
```

```python
fig, axes = plt.subplots(2, 1, figsize=(14, 12))
# 以下不會調用 plt，因為這個庫太大了
axes[0].plot(x, y1, label="sqrt") # axes[0]表示第一張圖
axes[0].plot(x, y2, label="sin")

axes[0].set_title('sqrt(x) vs sin(x)', fontsize=18)
axes[0].legend()

# axes[1] 表示第二張圖
axes[1].plot(x, y1, label="sqrt")
axes[1].plot(x, y3, label="cos")

axes[1].set_title('sqrt(x) vs cos(x)', fontsize=18)
axes[1].legend()

```

```python
fig, axes = plt.subplots(2, 1, figsize=(14, 12))
# 以下不會調用 plt，因為這個庫太大了
axes[0].plot(x, y1, label="sqrt") # axes[0]表示第一張圖
axes[0].plot(x, y2, label="sin")

axes[0].set_title('sqrt(x) vs sin(x)', fontsize=18)
axes[0].legend()

# axes[1] 表示第二張圖
axes[1].plot(x, y1, label="sqrt")
axes[1].plot(x, y3, label="cos")

axes[1].set_title('sqrt(x) vs cos(x)', fontsize=18)
axes[1].legend()

axes[1].set_xlabel('x axis', fontsize=14, fontfamily="serif", fontstyle="italic")
axes[1].set_ylabel('y axis', fontsize='x-large', fontstyle='oblique')

#設置橫坐標，equal表示 x 與 y 軸的比例一樣
axes[1].set_aspect('equal')

# 不論 x 範圍有多大，可以這樣設定限制他(與上圖比較，可以發現 x 的顯示範圍從 0 - 20 變成 0 - 16)
axes[1].set_xlim(0, 16)

# 或者可以訂製座標上面顯示的數，要多少間隔顯示一個(下面是間距為 1 的設置)
start, end = axes[1].get_xlim()
axes[1].xaxis.set_ticks(np.arange(start, end, 1))

# 現在改變 ticks顯示的樣子
axes[1].xaxis.set_tick_params(rotation=45, labelsize='large', color='b')

# 進一步地，示範讓 y 軸的 ticks 顯示在右邊
axes[1].yaxis.tick_right()

# 顯示 minor ticks(看圖就知道啥意思)
axes[1].minorticks_on()

# 還可以對網格線做設定: grid(which 表示要 major 還是 minor, axis 表示經線還是緯線, color 表示要顯示的顏色)
axes[1].grid(which="major", axis="both", color='r')
axes[1].grid(which='minor', axis="both")

# 對於子圖彼此關係的設定，首先可以嘗試一齊向左對齊 (預設是置中對齊)
# 注意這裡不是用 left、right，而是東西南北向表示！
# 再注意裏頭選擇的參數為大寫 C, SW, S, SE, E, NE, N, NW, W 其中一個
for ax in axes:
    ax.set_anchor("W")
    
# 還有可以控制各個子圖的設置，注意要用subplots才不會報錯
# hspace表示兩張圖的上下間隔設定
plt.subplots_adjust(hspace=0.3)

plt.show()
```
